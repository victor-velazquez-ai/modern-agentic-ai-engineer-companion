```python
# The branches, as the book lists them, each tied to its Part(s) and the repo route
# (the learn/ folder prefix) you'd work. This is reference data — don't edit it;
# you'll edit your RATINGS in the next cell.
from dataclasses import dataclass


@dataclass(frozen=True)
class Branch:
    name: str
    part: str        # the book Part(s), matching §2.2
    repo_route: str  # where to work in this repo when this branch is shaky/new


BRANCHES = [
    Branch("Foundations", "Part II",        "learn/part-02-*"),
    Branch("The model",   "Part III",       "learn/part-03-llm-substrate/"),
    Branch("Agents",      "Parts IV–V",     "learn/part-04-building-blocks-of-agents/ + part-05-*"),
    Branch("Quality",     "Part VI",        "learn/part-06-*"),
    Branch("Backend",     "Part VII",       "learn/part-07-*"),
    Branch("Cloud",       "Part VIII",      "learn/part-08-*"),
    Branch("Frontend",    "Part IX",        "learn/part-09-*"),
    Branch("Production",  "Parts X–XI",     "learn/part-10-* + part-11-*"),
    Branch("Frontiers",   "Part XII",       "learn/part-12-*"),
    Branch("Career",      "Part XIII",      "learn/part-13-*"),
]

print(f"{len(BRANCHES)} branches loaded — the ten from §2.2.")
for b in BRANCHES:
    print(f"  {b.name:12s} -> {b.part}")
```

## 🔧 Fill in your honest ratings

Rate each branch with exactly one of: `"solid"` (could teach it), `"shaky"` (use it but with gaps), or `"new"` (little or no experience). **Be ruthless** — the goal is to find blind spots, not to feel good (§2.3).

Edit the values in `RATINGS` below. Every branch starts at `"new"` so you can't skip one by accident.

```python
# ✍️ EDIT ME. Change each value to "solid", "shaky", or "new".
# This is the only cell you have to change to make the worksheet yours.
RATINGS = {
    "Foundations": "new",
    "The model":   "new",
    "Agents":      "new",
    "Quality":     "new",
    "Backend":     "new",
    "Cloud":       "new",
    "Frontend":    "new",
    "Production":  "new",
    "Frontiers":   "new",
    "Career":      "new",
}

VALID = {"solid", "shaky", "new"}


def validate(ratings: dict) -> None:
    """Fail fast and friendly if a rating is missing or mistyped."""
    expected = {b.name for b in BRANCHES}
    missing = expected - set(ratings)
    extra = set(ratings) - expected
    if missing:
        raise ValueError(f"Missing rating(s) for: {sorted(missing)}")
    if extra:
        raise ValueError(f"Unknown branch name(s): {sorted(extra)} (check spelling)")
    bad = {k: v for k, v in ratings.items() if v not in VALID}
    if bad:
        raise ValueError(f"Each rating must be one of {sorted(VALID)}; got {bad}")
    print("✓ all ten branches rated with a valid value.")


validate(RATINGS)
```

## Turn ratings into a route

Now the payoff. The next cell reads your ratings and prints a plan: for each *shaky* or *new* branch it names the Part(s) and the `learn/` folders to work **fully** (do every 🔧 Build); for *solid* branches it tells you to **skim for the senior/architect lenses** (§2.3). This is your reading order, generated from your own honesty.

```python
def build_route(ratings: dict) -> None:
    """Print a branch-by-branch plan from solid/shaky/new ratings."""
    by_name = {b.name: b for b in BRANCHES}
    work_fully, skim = [], []
    for b in BRANCHES:  # iterate in book order, not dict order
        (skim if ratings[b.name] == "solid" else work_fully).append(b)

    print("=" * 64)
    print("YOUR PERSONAL ROUTE THROUGH THE BOOK + REPO")
    print("=" * 64)

    print(f"\n🔧 WORK FULLY (do every Build) — {len(work_fully)} branch(es):")
    if not work_fully:
        print("  (none — you rated everything solid; skim for the lenses, then go build.)")
    for b in work_fully:
        tag = ratings[b.name].upper()
        print(f"  [{tag:5s}] {b.name:12s} {b.part:11s}  {b.repo_route}")

    print(f"\n🎯 SKIM FOR THE SENIOR/ARCHITECT LENSES — {len(skim)} branch(es):")
    if not skim:
        print("  (none yet — nothing is solid. That's fine; that's why you're here.)")
    for b in skim:
        print(f"  [SOLID] {b.name:12s} {b.part:11s}  {b.repo_route}")

    weakest = [b.name for b in BRANCHES if ratings[b.name] == "new"]
    print("\n" + "-" * 64)
    print(f"Weakest (rated 'new'): {weakest or '— none'}")
    print("🔮 Compare that to the branch you predicted earlier. Surprised?")
    print("Suggested start: the lowest-numbered Part among your 'work fully' branches.")


build_route(RATINGS)
```

**What you just saw.** Your honest ratings became a concrete order of operations — which Parts to grind through doing every Build, and which to skim for the architect-level judgment. That mapping is the whole point of §2.3: the assessment *is* the plan.

## ⚠️ Pitfall: rating yourself solid to feel good

The most common way this worksheet fails is **flattery** — rating a branch *solid* because you've *touched* it, not because you could *teach* it. "I've called an LLM API" is not the same as *solid* on **The model** (do you know how tokenization drives cost, or what sampling knobs do under latency?). Rating up moves a branch to "skim," so you skip exactly the chapters that would have closed a real gap — and the gap surfaces later, in production, where it's expensive.

**The fix:** use the chapter's bar literally. *Solid* = "could teach it." When in doubt, rate **down** (`shaky`, not `solid`). Re-scoring every few Parts (below) catches over-optimism cheaply; a missed gap in Part X does not.

## Re-score: watch *new* → *solid*

§2.3 says to revisit the assessment every few Parts — watching *new* turn into *solid* is the most motivating progress signal you have. Drop a dated snapshot of your ratings into `HISTORY` as you go; the cell shows movement since your first pass.

```python
# ✍️ Append a new dated copy of RATINGS each time you re-score (after every few Parts).
# Keep the first entry as your baseline; add later ones below it.
HISTORY = [
    {"date": "2026-06-20", "ratings": dict(RATINGS)},  # baseline = your first pass
    # {"date": "2026-07-15", "ratings": {"Foundations": "solid", ...}},
]

RANK = {"new": 0, "shaky": 1, "solid": 2}


def show_progress(history: list) -> None:
    """Compare the latest snapshot to the baseline and report what improved."""
    if len(history) < 2:
        print("Only a baseline so far. Re-run this after a few Parts to see movement.")
        return
    first, last = history[0]["ratings"], history[-1]["ratings"]
    print(f"Progress from {history[0]['date']} to {history[-1]['date']}:")
    improved = 0
    for b in BRANCHES:
        a, z = first[b.name], last[b.name]
        if RANK[z] > RANK[a]:
            print(f"  ↑ {b.name:12s} {a} -> {z}")
            improved += 1
    print(f"\n{improved} branch(es) leveled up. That's the signal §2.3 wants you to chase.")


show_progress(HISTORY)
```
