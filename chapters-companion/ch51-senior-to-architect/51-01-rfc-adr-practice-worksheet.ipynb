## 🧠 Why this matters

The jump from senior to architect is a change of *instrument*: your output stops being the systems you build and becomes the systems you cause to be built well, by other people (§51 intro). The two things that make that real are **second-order thinking** — tracing what a decision makes *people* do, not just what it does to the system — and **driving decisions through documents and rooms** instead of in your head (§51.1, §51.3).

None of that is executable. A cell can't teach you to pre-align a skeptical stakeholder in the hallway before the meeting, or to answer an exec's risk question in their currency instead of dumping pipeline internals on them. So this notebook makes you **decide, write, and socialize** on a *real* decision you face — and produces two durable artifacts (an RFC and an ADR) you keep using at work.

## 🧠 Player → game designer: trace the second-order effects

A senior asks "is this design good?" An architect asks: *given these rules, boundaries, and incentives, what will every player naturally do — and is that the system we want?* (§51.1). **First-order effects** are what a decision does; **second-order effects** are what the decision makes *people* do. The classic traps in the chapter are behavioral, not technical: per-token pricing makes teams quietly avoid your platform; a heavyweight review process makes risky changes get split into many “small” ones.

And in agentic systems you are designing games with *non-human players in them* too — so trace the agent's behavior as well: what does it do when a tool times out? What does your cost cap incentivize the orchestrator toward? (§51.1 senior lens.)

## Multiplier check: the two-week-vacation test (§51.2)

## 🔧 Write the RFC — problem, forces, honest options, recommendation (§51.3)

Decisions in organizations don't happen in your head; they happen in **documents and rooms**. Step one of the loop is to *frame the decision in writing first*: an RFC that states the problem, the forces, **two or three real options with honest trade-offs**, and a recommendation. Writing it forces the thinking; circulating it scales the thinking (§51.3).

Fill in the template below in your own words. It lifts directly into [`../../templates/system-design-doc/`](../../templates/system-design-doc/) — the RFC/options + trade-off scaffold from Ch 27.

### ✍️ Your RFC

```markdown
# RFC: <one-line decision, stated as a question>

## Status
Draft — <date> — author: <you>

## Problem
<What forces the decision now? Who is blocked, what breaks, or what opportunity
closes if we don't choose. Keep it to the decision — not a system tour.>

## Forces (ranked)
<The real constraints and the top quality attributes for THIS decision: team size,
deadline, scale, reversibility, skills. Rank them — an unranked list hides the trade-off.>

## Options (two or three, each with HONEST trade-offs)
### Option A — <name>
+ <what it gives>   - <what it costs / what you give up>
### Option B — <name>
+ <what it gives>   - <what it costs>
### Option C — <name, optional>
+ <what it gives>   - <what it costs>

## Recommendation
<Lead with the conclusion: "We should do B." Then the two or three reasons, and the
evidence (eval data, costs, the ranked forces). Name the SECOND-ORDER effect you
expect this to create in people's behavior — and why it's acceptable.>
```

> **An RFC with one option is a memo, not a decision (§51.3).** If you can't write an honest `-` line for your recommended option, you haven't found its real cost yet.

## Pre-wiring plan: decisions are made in the hallway (§51.3)

Step two of the loop: **pre-align before the meeting.** Walk the document past key stakeholders *individually*, before any group discussion. You surface objections while they're cheap, give people time to think and to feel heard, and arrive with most of the room already nodding. *Decisions are made in the hallway; meetings ratify them.*

This is the chapter's reframe: stakeholder alignment is **distributed-systems engineering applied to people** — stakeholders are nodes with partial information, different objectives, and unreliable message delivery. Pre-wiring is *replication*; the ADR is the *durable log* (§51.3 key idea).

### ✍️ Your pre-wiring plan

List the stakeholders, the objection each will raise, and the **one-on-one order** to walk the doc — walk the most skeptical / most influential *first*, so the group meeting is a ratification, not a fight.

| Order | Stakeholder | Their objective | The objection they'll raise | What addresses it before the meeting |
|---|---|---|---|---|
| 1 | _<name / role>_ | _<what they optimize for>_ | _<their likely objection>_ | _<the answer / change you bring them>_ |
| 2 | _<…>_ | _<…>_ | _<…>_ | _<…>_ |
| 3 | _<…>_ | _<…>_ | _<…>_ | _<…>_ |

> **Replication, not politics (§51.3).** If a stakeholder first hears your proposal *in the meeting*, you've delivered a message over an unreliable channel with no ack — expect a packet drop.

## 🔧 Record the ADR — the durable log (§51.3)

Step four: **record it in an ADR and move on.** The why, the rejected options, the revisit conditions — this is what stops the organization re-litigating the same question every quarter. Same ADR-014 shape you used in Ch 27 (**Status / Context / Decision / Consequences**, plus the alternatives you rejected). It lifts into [`../../templates/adr-template/`](../../templates/adr-template/).

Here is the book's **ADR-014** as the reference shape — your RFC's chosen option becomes this ADR's `Decision`:

```markdown
# ADR-014: Use a modular monolith for the agent platform

## Status
Accepted — 2026-06-09

## Context
Two engineers, pre-launch, requirements still shifting. We need fast iteration
and one simple deploy, but also clean seams so we can extract services later.

## Decision
Build a modular monolith: strict module boundaries (agents, retrieval, billing)
with no cross-module imports except through defined interfaces; one deployable.

## Consequences
+ Fast to build, simple to run and debug; easy to refactor boundaries early.
- Requires discipline to keep modules separate; shared DB needs care.
- Revisit and extract services when a module needs independent scaling or a team.
```

> The ADR is your **durable log** (§51.3). The next person who asks “why didn't we just use microservices?” reads this instead of reopening the meeting.

## Altitude drill: one decision, three rooms (§51.4)

The most visible mark of an architect is explaining the **same** system at three levels — and *detecting which one the room needs*. Write your decision three ways, and **lead with the conclusion** every time (§51.4):
- **Exec — one-sentence business outcome.** Speak in money, time, risk, optionality. “Tech debt in the orchestrator” moves no one; “every new agent capability now costs roughly double to ship, and that gap is widening” moves budgets.
- **Fellow leads — the boundary and the trade-off.** What boundary moves, what you traded for what.
- **Engineers — the mechanism.** How it actually works.

### ✍️ Three framings of the *same* decision (conclusion first, each time)

| Audience | Altitude | Your framing (start with the recommendation) |
|---|---|---|
| Executive | One-sentence business outcome — money / time / risk / optionality | _<“We should …, because it cuts … / de-risks ….”>_ |
| Fellow leads | Boundary + trade-off | _<“We move the … boundary; we gain … at the expense of ….”>_ |
| Engineers | Mechanism | _<“Concretely: … calls … through …; the rule is ….”>_ |

> **Lead with the conclusion (§51.4).** Engineers love building *to* the conclusion like a proof; executives experience that as burying the answer. Conclusion-first also forces *you* to actually have one. Treat this as **context engineering for humans**: the right information, at the right altitude, with irrelevant detail pruned (§51.4 mental model).

## ⚠️ Pitfall: answering the exec with pipeline internals

The failure that marks someone *not yet* at this altitude: an executive asks a **risk** question (“what's the chance this slips the launch?”) and the answer comes back as retrieval-pipeline internals — chunk sizes, reranker latency, token budgets. That isn't rigor; it's a failure to **route the message** to the room (§51.4). It's the same mistake as carefully curating an agent's context window and then dumping forty unfiltered slides on a VP: you mastered the skill and missed its point.

### ✍️ Catch yourself doing it

- Write the **wrong** answer to your exec — the one that reaches for mechanism: _<the pipeline-internals version>_
- Now rewrite it as **outcome + risk** in their currency: _<money / time / risk / optionality>_

## 🎯 Senior lens

Everything in this worksheet is the human edge being *exercised* (§51 intro; compounds with Ch 27's thesis that architecture is the human edge in the agentic era). Three judgment calls a senior makes here that a junior misses. **First, second-order beats first-order:** the junior ships the locally-good design; the architect asks what the rule makes every player *do* — and treats the agent in the loop as one of those players, so cost caps and timeouts become correctness concerns, not just leadership niceties (§51.1). **Second, the multiplier dominates the addend:** the most expensive thing a new lead does is keep being the team's best individual contributor; the vacation test is how you notice (§51.2). **Third, the doc is the decision:** seniors-and-above are paid for decisions, and decisions live in pre-wired RFCs and durable ADRs, not in your head — the engineer who frames stakeholder alignment as distributed systems stops resenting it and starts getting good at it (§51.3).

## Recap

- **Player → game designer (§51.1):** trace a change's *second-order* effects — what it makes people *and agents* do — not just its first-order effect on the system.
- **Multiplier, not addend (§51.2):** the two-week-vacation test reveals the bottleneck; escape the hero-engineer trap by delegating *outcomes*, not tasks.
- **The decision-driving loop (§51.3):** frame it in an **RFC** with honest options → **pre-wire** stakeholders one-on-one → force an explicit call (named maker + date + default) → record an **ADR**.
- **Pre-wiring is replication; the ADR is the durable log (§51.3):** stakeholder alignment is distributed-systems engineering applied to people.
- **Altitude switching (§51.4):** explain the *same* decision at exec / lead / engineer altitude, lead with the conclusion, and never answer a risk question with pipeline internals.
