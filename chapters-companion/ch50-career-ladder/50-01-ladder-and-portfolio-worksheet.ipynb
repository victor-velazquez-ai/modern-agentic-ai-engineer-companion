## 🧠 The ladder is *scope of ownership*, not years or tech

Each level is defined not by years served or technologies known, but by the **scope of problem you can be handed without supervision** — and by how much of your impact flows *through other people*. Two patterns matter: the **code share of your impact falls at every level** (a junior's impact is nearly all code; a principal's is nearly all decisions and other people's output), and the **senior → staff transition is the sharpest discontinuity** — it's where impact-through-others stops being optional.

Run the cell to see the §50.1 ladder. Then **fill in your self-locator** in the markdown cell that follows.

```python
# The §50.1 ladder, verbatim from the book's Table. Reference only — no judgment here yet.
ladder = [
    ("Junior",    "A well-defined task",            "Execute reliably with guidance; main output is your own learning rate."),
    ("Mid",       "A feature, end to end",          "Handed problems, not tasks; decompose them yourself and ship without hand-holding."),
    ("Senior",    "A system or product area",       "Own outcomes, not code. Make trade-offs, unblock others, be the quality bar for your area."),
    ("Staff",     "A problem spanning teams",       "Unit of work is the *initiative*: align several teams on a technical direction and land it."),
    ("Principal", "Org-level technical direction",  "Shape what the org builds and how — strategy, standards, the hardest cross-cutting calls."),
    ("Architect", "System coherence at scale",      "Own the expensive decisions across a portfolio of systems, plus the people-alignment to make them stick."),
]

print(f"{'Level':<10} | {'Scope you own':<32} | What changes when you get there")
print("-" * 110)
for level, scope, change in ladder:
    print(f"{level:<10} | {scope:<32} | {change}")

print("\nNote: the CODE share of your impact falls at every level; the Senior->Staff jump is the sharpest.")
```

## The skill/impact matrix: how promotions really work (§50.2)

The mechanism almost nobody explains: **skill and impact are different axes, and promotions are paid out on impact** — the difference your work made to something the org cares about (revenue, cost, reliability, the speed of other teams). The two correlate far more loosely than feels fair.

| | **Low visibility of impact** | **High visibility of impact** |
|---|---|---|
| **High skill** | The **"hidden gem"**: brilliant, frustrated, perpetually passed over. The most common failure mode of strong engineers. | The **promotion track**. Skill applied to problems leadership cares about, with results legible to them. |
| **Lower skill** | Stagnation; usually managed out over time. | Short-lived: visibility without substance collapses at the first hard project. |

Promotions live in the **top-right** cell. Most strong engineers who feel stuck are in the **top-left** — and the fix is *not more skill*. Run the cell to lay out your recent work, then rate it.

```python
# A scaffold to rate your recent work on BOTH axes. Replace the sample rows with your own
# last 3-5 substantial pieces of work. Skill 1-5 (how hard/senior the work was);
# Visibility 1-5 (could someone with promotion power see and understand the impact?).
recent_work = [
    # (description, skill_1to5, visibility_1to5)
    ("Cut agent p95 latency 40% by fixing a retrieval N+1", 4, 2),   # sample: a HIDDEN GEM
    ("Wrote the launch eval harness everyone now relies on", 5, 2),  # sample: a HIDDEN GEM
    ("Demoed cost dashboard at the all-hands; VP cited it",  3, 5),  # sample: promotion track
]


def quadrant(skill, vis):
    hi_skill, hi_vis = skill >= 4, vis >= 4
    if hi_skill and hi_vis:
        return "PROMOTION TRACK (top-right)"
    if hi_skill and not hi_vis:
        return "HIDDEN GEM (top-left) <- the fix is visibility, not more skill"
    if not hi_skill and hi_vis:
        return "visibility w/o substance (short-lived)"
    return "stagnation"


print(f"{'Work':<52} | Skill | Vis | Quadrant")
print("-" * 100)
for desc, skill, vis in recent_work:
    print(f"{desc:<52} | {skill:^5} | {vis:^3} | {quadrant(skill, vis)}")

gems = [d for d, s, v in recent_work if s >= 4 and v < 4]
print(f"\nHidden-gem items flagged: {len(gems)}")
for g in gems:
    print(f"  - {g}")
```
