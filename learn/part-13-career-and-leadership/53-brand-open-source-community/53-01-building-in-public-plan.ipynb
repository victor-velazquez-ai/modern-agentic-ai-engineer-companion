## Learning exhaust: never create from nothing

The rule that makes publishing sustainable: **treat content as *learning exhaust***. You are already learning constantly; publishing captures a byproduct, not a second job. Never create from nothing — write what you **debugged, built, measured, or decided** this month. And use your perishable advantage: *write what you just learned, for yourself one month ago*, while you still remember what was confusing (experts can't teach beginners well because they've lost that memory; you have it fresh).

The three genres that work, from §53.1:

- **Debugging stories** — "our agent looped for forty minutes: a trace-level post-mortem." War stories carry judgment, and judgment is what hiring managers scan for.
- **Honest evaluations with numbers** — "we tested three chunking strategies; here's the eval data." Rare, and cited.
- **Explanations at the right altitude** — take one thing this book taught you and explain it in your own words and examples (also the best comprehension test you'll take).

List four real things from the last month or two; each one is a post you already have the material for.

```python
# ✍️ EDIT ME — four real things from your recent work. Each is a post you can already write.
@dataclass
class ExhaustItem:
    what: str     # the real thing you debugged / built / measured / decided
    genre: str    # "debugging story" | "honest eval with numbers" | "explanation at altitude"


EXHAUST = [
    ExhaustItem("...", "debugging story"),
    ExhaustItem("...", "honest eval with numbers"),
    ExhaustItem("...", "explanation at altitude"),
    ExhaustItem("...", "debugging story"),
]

GENRES = {"debugging story", "honest eval with numbers", "explanation at altitude"}


def pick_next(items: list) -> ExhaustItem:
    """Choose the next artifact: the first real (non-placeholder) item, else nudge."""
    for i in items:
        if i.genre not in GENRES:
            print(f"⚠️  '{i.genre}' isn't one of the three book genres: {sorted(GENRES)}")
    real = [i for i in items if not i.what.strip().startswith(".")]
    if not real:
        print("⚠️  All four are placeholders. You did real work this month — name it. "
              "If you truly can't, that's the signal: ship something worth writing about.")
        return items[0]
    nxt = real[0]
    print(f"Next artifact -> [{nxt.genre}] {nxt.what}")
    print(f"You have {len(real)} queued. That's months of posts from work you already did.")
    return nxt


NEXT_POST = pick_next(EXHAUST)
```

## Open source: pick exactly one tier

Open source is the most *verifiable* reputation channel — anyone can read your code, your reviews, and how you handle disagreement in an issue thread. But "do open source" is bad advice without a strategy. Three escalating tiers, each a different cost/return (§53.2). **Pick one — honestly — for the next two years:**

1. **Focused contributions.** A dozen thoughtful PRs to *one* tool you actually use (an agent framework, an eval library, an MCP server), building trust with its maintainers — worth more than fifty drive-by typo fixes. Reputation in open source is *per community*, and compounds where you stay.
2. **Owning a niche.** Build and maintain the small, sharp tool the ecosystem is missing. A modestly-used project, maintained well, demonstrates the *full* engineering loop — API design, docs, releases, breaking-change judgment, saying no — which is staff-level evidence no employer's codebase can show publicly.
3. **Maintainership of something significant.** Highest tier, highest cost. Real authority, real burnout risk. Enter it deliberately or not at all.

```python
# ✍️ EDIT ME — choose ONE tier and your first three concrete actions.
TIER = 1   # 1 = focused contributions | 2 = owning a niche | 3 = significant maintainership

FIRST_THREE_ACTIONS = [
    "...",   # e.g. "fix the open issue I hit in <tool> this week"
    "...",
    "...",
]

TIERS = {
    1: ("Focused contributions",
        "Low cost, compounding return. Reputation is per-community — stay in ONE.",
        "Best first move with a day job: the tools you already use generate the PRs."),
    2: ("Owning a niche",
        "Medium cost, staff-level evidence. Shows the FULL loop: API, docs, releases, saying no.",
        "Pick a real gap you keep hitting; a small sharp tool beats a sprawling unfinished one."),
    3: ("Significant maintainership",
        "Highest cost AND highest burnout risk. Real authority — enter deliberately or not at all.",
        "Only if you can sustain it for years; otherwise tier 1 or 2 is the honest choice."),
}


def choose_tier(tier: int, actions: list) -> None:
    if tier not in TIERS:
        print("⚠️  Pick exactly one tier: 1, 2, or 3.")
        return
    name, cost_return, note = TIERS[tier]
    print(f"Tier {tier}: {name}")
    print(f"  cost/return: {cost_return}")
    print(f"  note: {note}")
    real = [a for a in actions if not a.strip().startswith(".")]
    if len(real) < 3:
        print("⚠️  Name three CONCRETE first actions (not 'contribute more'). "
              "Specific actions get done; intentions don't.")
    else:
        print("  first three actions:")
        for a in real:
            print(f"    - {a}")


choose_tier(TIER, FIRST_THREE_ACTIONS)
```

## 🎯 Senior lens

Read your open source as a **demonstration of the judgment layer**. Anyone can claim "production experience" in an interview; a public repo with **ADRs**, an **eval suite**, a tidy **changelog**, and graceful issue-thread conduct *proves* it. As AI makes code generation cheap, this shifts further in your favor: what a repo uniquely shows is no longer that you *can write code* — it's how you **decide, review, and maintain**.

Curate accordingly. Make the README of your flagship project explain the **why**, keep ADRs and an eval suite visible, and **archive the abandoned experiments** so the signal isn't buried in noise. Your `capstone/` (Appendix C), once operated and documented, is exactly this flagship — the source of the debugging stories and eval numbers you publish. The curation *is* the brand; the code is just the substrate.

## ⚠️ Pitfall: optimizing for reach instead of the record

The most common way this whole plan fails is **chasing the wrong metric**. Reach — followers, likes, daily posts — is easy to measure and seductive, so people optimize it: performative volume, hot takes, recycled platitudes. But the audience that changes your career (senior engineers, hiring managers, maintainers) actively *discounts* that signal. You build reach and *lose* reputation.

The mirror-image failure is **perfectionism** — hoarding a six-weeks-polished post that never ships, so the record stays empty.

**The fix:** optimize for the *verifiable record*, not the vanity metric. One substantive artifact a month — a real debugging story, an eval with numbers, a well-maintained PR thread — compounds into authority. A year of daily platitudes compounds into nothing. When you feel the pull toward volume *or* toward endless polishing, return to the one rule you wrote two cells up.

## Networking that compounds

Strip the slime off the word: networking is the long-term accumulation of people who know what you can do and think well of you. It matters because the best opportunities — early-stage roles, architect positions, founding teams — are filled through **trust before they are ever posted**. Granovetter's *The Strength of Weak Ties* (1973) found opportunities flow disproportionately through **weak ties**: not close friends (who know what you know) but acquaintances in adjacent circles who bridge into networks you can't see.

The compounding moves are unglamorous (§53.3):

- **Give first, specifically** — a thoughtful answer in a community Slack, a code review on a stranger's project, an intro between two people who should meet. Givers are remembered; extractors are filtered.
- **One community, deeply** — the working group, the local AI-engineering meetup, the framework's contributor channel. Showing up repeatedly is the whole trick; familiarity is trust's raw material.
- **Maintain, lightly** — a two-line "saw this and thought of your project" twice a year keeps a tie alive.

```python
# ✍️ EDIT ME — your give-first plan and the ONE community you'll engage deeply.
GIVE_FIRST = [
    "...",   # a specific unprompted useful act: answer / review / intro
    "...",
    "...",
]
DEEP_COMMUNITY = "..."           # the ONE community you'll show up in repeatedly
LIGHT_TOUCHES_PER_YEAR = 2       # "saw this, thought of your project" cadence

WEEKLY_NUDGES = [
    "Answer one question in your deep community this week.",
    "Review one stranger's PR or repo, specifically and kindly.",
    "Make one intro between two people who should know each other.",
]


def networking_plan(gives: list, community: str, touches: int) -> None:
    real = [g for g in gives if not g.strip().startswith(".")]
    if not real:
        print("⚠️  Give-first list is empty. The unit of network-building is the "
              "unprompted useful act — name three you can do this month.")
    else:
        print("Give-first (no immediate return expected):")
        for g in real:
            print(f"  - {g}")
    if community.strip().startswith("."):
        print("⚠️  Pick ONE community to go deep in. Ten joined-and-lurked < one engaged.")
    else:
        print(f"Deep community: {community}  (show up repeatedly — that's the trick)")
    print(f"Light-maintenance: {touches}x/year 'saw this, thought of your project' notes.")
    # The one seeded nudge — deterministic via random.seed(53) in Setup.
    print(f"This week's give: {random.choice(WEEKLY_NUDGES)}")


networking_plan(GIVE_FIRST, DEEP_COMMUNITY, LIGHT_TOUCHES_PER_YEAR)
```

## The highest-return move: tell the three people whose work yours builds on

The networking move with the highest return-per-awkwardness (§53.3 💡): when you publish something, send it **directly to the three people whose work it builds on**, with **one specific sentence** on what you took from theirs. It's honest, it's flattering *because* it's specific, and it reliably starts real correspondence with people otherwise out of reach.

Draft that message **now**, for the artifact you queued earlier (`NEXT_POST`).

```python
# ✍️ EDIT ME — name the three people and what, specifically, you took from each.
INSPIRATIONS = [
    ("...", "..."),   # (person/handle, the one specific thing their work gave you)
    ("...", "..."),
    ("...", "..."),
]


def draft_outreach(post, people: list) -> None:
    """Assemble the highest-return message — printed only, never sent (no network)."""
    title = post.what if not post.what.strip().startswith(".") else "<your next post>"
    print(f"When '{title}' ships, send each of these — privately, one specific sentence:")
    print()
    for who, took in people:
        if who.strip().startswith(".") or took.strip().startswith("."):
            print("  ⚠️  Fill in a real person AND the specific thing you took. "
                  "Vague flattery reads as spam; specificity is the whole point.")
            continue
        msg = (f'Hi {who} — I just published "{title}". '
               f"It builds directly on your work: {took}. Thank you for it.")
        print(textwrap.fill(msg, width=78, initial_indent="  ", subsequent_indent="    "))
        print()
    print("NOTE: this worksheet only DRAFTS the message. Sending happens by you, "
          "outside the repo — no network, no posting here.")


draft_outreach(NEXT_POST, INSPIRATIONS)
```

## 📋 Self-audit: the §53 checklist, on a cadence

This is the chapter's closing checklist (§53) turned into a *recurring* audit — the point of the whole worksheet is that you re-run it, not that you answer it once. Mark each item honestly; the cell scores you and names what's slipping. Re-run it **monthly** and **quarterly**, the way the questions are phrased.

```python
# ✍️ EDIT ME — answer each §53 checklist item True/False, honestly, for THIS month/quarter.
@dataclass
class CheckItem:
    question: str
    done: bool
    cadence: str   # "month" or "quarter"


CHECKLIST = [
    CheckItem("Published one substantive artifact (post, talk, repo) in the last MONTH?", False, "month"),
    CheckItem("Published one substantive artifact in the last QUARTER?", False, "quarter"),
    CheckItem("One open-source project where maintainers recognize your name?", False, "quarter"),
    CheckItem("Flagship public repo shows JUDGMENT — ADRs, evals, docs — not just code?", False, "quarter"),
    CheckItem("A stranger could find your niche and work in ONE search, consistent across platforms?", False, "quarter"),
    CheckItem("Helped three people this quarter with NO immediate return?", False, "quarter"),
    CheckItem("Three people who would vouch for you?", False, "quarter"),
]


def self_audit(items: list, today: date) -> None:
    done = sum(1 for i in items if i.done)
    print(f"§53 self-audit — {today.isoformat()}  ({done}/{len(items)} green)")
    print()
    for i in items:
        mark = "✅" if i.done else "⬜"
        print(f"  {mark} [{i.cadence:7s}] {i.question}")
    slipping = [i for i in items if not i.done]
    print()
    if not slipping:
        print("All green. The plan is holding — keep the cadence; don't add volume.")
    else:
        worst = next((i for i in slipping if i.cadence == "month"), slipping[0])
        print(f"Most urgent slip: {worst.question}")
        print("Fix the MONTH-cadence items first — they're the leading indicator. "
              "Re-run this cell next month with a fresh date.")


self_audit(CHECKLIST, date.today())
```
